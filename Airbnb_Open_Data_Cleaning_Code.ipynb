```python
"""1. Deleting redundant columns.
2. Renaming the columns.
3. Dropping duplicates
4. Rename the NaN values from the datasets.
5. Cleaning individual columns
6. Check for some more Transformations."""
```

```python
col_to_keep = ['NAME', 'host id', 'host_identity_verified', 'host name',
       'neighbourhood group', 'neighbourhood', 'lat', 'long', 'country',
       'country code', 'instant_bookable', 'cancellation_policy', 'room type',
       'Construction year', 'price', 'service fee', 'minimum nights',
       'number of reviews', 'last review']
col_to_drop = ['reviews per month',
       'review rate number', 'calculated host listings count',
       'availability 365', 'house_rules', 'license', 'id']
```

```python
print(len(col_to_keep))
print(len(col_to_drop))
```

# 1st Method - Filtering Data

```python
df = data[col_to_keep]
```

```python
print(data.shape) #the original dataframe is not change
print(df.shape) #the new appointed dataframe of 'df'
```

# 2nd Method - Dropping Columns

```python
data.drop(columns = col_to_drop, inplace = True) # inplace makes the changes in original dataset
```

```python
data.shape
```

# Renaming

```python
data.rename(columns = {"NAME": "Name"}, inplace = True) # Method 1
```

```python
new_col_names = []
for i in data.columns:
    new_col_names.append(i.upper())  # Method 2 using "upper" or "lower"
```

```python
new_dataframe_to_modify = data.copy()
```

```python
new_dataframe_to_modify.columns = new_col_names
```

# Cleaning Individual columns

```python
data["host_identity_verified"].str.upper()
```

```python
data
```

```python
data["host_identity_verified"] = data["host_identity_verified"].str.upper()
```

```python
data["instant_bookable"] = data["instant_bookable"].apply(lambda x: 1 if x==True else 0)
```

```python
data.reset_index(inplace = True)
```

```python
type(data["price"][15])
```

```python
data["price"].head(3)
```

```python
data["price"] = data["price"].str.replace("$", "")
```

```python
data["price"] = data["price"].str.replace(",", "")
```

```python
data["price"] = data["price"].str.replace(" ", "")
```

```python
data["price"][15]
```

```python
data["price"] = data["price"].astype(int)
```

```python
data.to_csv("Cleaned_Airbnb_Open_Data.csv", index = False)
```
