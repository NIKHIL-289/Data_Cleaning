# 1. Deleting redundant columns

```python
data.columns
```

```python
columns_to_keep = ['Age', 'Gender', 'Marital Status', 'Occupation', 'Monthly Income', 'Educational Qualifications', 'Family size', 'Output', 'Feedback']
columns_to_drop = ['latitude', 'longitude', 'Pin code', 'Unnamed: 12']
```

## Method 1

```python
modified_data = data[columns_to_keep].copy() 
""" .copy() is used becaue modified_data is probably a "view" (or slice) 
of another DataFrame, not a fully independent copy. When you do in-place modifications (inplace=True) on such views, 
Pandas cannot guarantee that the original DataFrame will or will not be affected, which is why it shows this warning"""
```

```python
modified_data
```

## Method 2

```python
refined_data = data.drop(columns = columns_to_drop)
```

```python
refined_data 
```

# 2. Renaming the columns.

```python
modified_data.columns = modified_data.columns.str.replace(" ", "_")
```

```python
modified_data.columns
```

```python
modified_data.rename(columns = {"Monthly Income": "Monthly Salary"}, inplace = True) 
"""" if we would not have used .copy() before this code would 
have given a warning and would have not changed modified_data accordingly """
```

```python
new_modified_data_columns = []
for i in modified_data.columns:
    new_modified_data_columns.append(i.upper())
```

```python
modified_data.columns = new_modified_data_columns 
```

```python
modified_data.columns
```

```python
modified_data
```

```python
modified_data.isna()
```

```python
# Anyways
modified_data.dropna(inplace = True)
```

# 5. Cleaning individual columns.

```python
modified_data
```

```python
modified_data["OUTPUT"] = modified_data["OUTPUT"].apply(lambda x: 1 if x=="Yes" else 0)
```

```python
modified_data["FEEDBACK"] = modified_data["FEEDBACK"].apply(lambda x: 1 if x=="Positive" else 0)
```

```python
modified_data["EDUCATIONAL_QUALIFICATIONS"] = modified_data["EDUCATIONAL_QUALIFICATIONS"].apply(lambda x: "PG" if x=="Post Graduate" else("GRAD" if x=="Graduate" else("PH.D" if x=="Ph.D" else("UNEDU"if x=="Uneducated" else "SCHOOL"))))
```

```python
modified_data
```

```python
unique_values = list(set(modified_data["EDUCATIONAL_QUALIFICATIONS"]))
print(unique_values)
```
