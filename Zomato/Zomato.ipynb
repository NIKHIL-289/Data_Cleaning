```python
"""Data Cleaning:

Deleting redundant columns.
Renaming the columns.
Dropping duplicates.
Remove the NaN values from the dataset
Cleaning individual columns.
Check for some more Transformations
"""
```

```python
columns_to_keep =['name', 'online_order', 'book_table', 'rate','dish_liked', 'approx_cost(for two people)'] 
columns_to_drop = ['url', 'address','votes','phone', 'location', 'rest_type', 'cuisines', 'reviews_list', 'menu_item', 'listed_in(type)', 'listed_in(city)']
```

# 1st way : Filtering the DATA

```python
data[columns_to_keep]
```

# 2nd way : Droping the COLUMNS

```python
data.drop(columns = columns_to_drop, inplace = True)
```

# Renaming the COLUMNS

```python
data.rename(columns = {"name":"NAME"})
```

```python
new_columns_names = []
for i in data.columns:
    new_columns_names.append(i.capitalize())
```

```python
new_columns_names
```

```python
data.columns = new_columns_names
```

# Dropping the DUPLICATES

```python
data.duplicated() # which line has duplicates returns True
```

```python
print(data.duplicated().sum()) # returns only the true valued number of rows
print(data.duplicated().value_counts()) # returns both the false valued and the true valued number of rows
```

```python
data.drop_duplicates(inplace = True) #
```

```python
data.duplicated().value_counts()
```

# Remove the NaN values

```python
data.isna().sum()
```

```python
data.dropna(inplace = True)
```

```python
data.isna().sum()
```

```python
data.shape
```

```python
data.reset_index(inplace = True)
```

```python
data
```

# Cleaning individual columns

```python
data["Rate"]
```

```python
data["Rate"] = data["Rate"].str.replace(" ", "")
```

```python
data["Rate"] = data["Rate"].str.replace("/", " out of ")
```

# Text to Int

```python
data["Online_order"] = data["Online_order"].apply(lambda x: 1 if x=="Yes" else 0)
```

```python
data["Book_table"] = data["Book_table"].apply(lambda x: 1 if x=="Yes" else 0)
```

```python
data.to_csv("Cleaned_zomato_data", index = False)
```
